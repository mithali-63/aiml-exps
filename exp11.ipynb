```python
import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split 
from sklearn.metrics import accuracy_score 
from sklearn.neighbors import KNeighborsClassifier 
import matplotlib.pyplot as pltfrom 
from mlxtend.plotting  import plot_decision_regions 
df = pd.read_excel("KNN_DS.xlsx") 
df.head() 
```

```python
X = df[["Height_cm", "Weight_kg"]] 
y = df["BMI_Class"] 
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=0)
knn = KNeighborsClassifier(n_neighbors=5) 
knn.fit(X_train, y_train) 
y_pred = knn.predict(X_test) 
accuracy = accuracy_score(y_test, y_pred) 
print("Accuracy:", accuracy)
```

```python
for i in range(len(X_test)):  
    print("Test data:", X_test.iloc[i], "| True label:", y_test.iloc[i], "| Prediction:", y_pred[i])  
    if y_test.iloc[i] == y_pred[i]:  
        print("Correct prediction")  
    else:  
        print("Wrong prediction")
```

```python
H = float(input("Enter height (in CM): ")) 
W = float(input("Enter weight (in KG): ")) 
predicted_label = knn.predict([[H, W]]) 
print("Predicted label:", predicted_label[0]) 
```

```python
knn.fit(X_train, y_train) 
plot_decision_regions(X_train.values, y_train.values, clf=knn, legend=2) 
plt.xlabel('Height (CM)')
plt.ylabel('Weight (KG)') 
plt.title('KNN Decision Boundaries') 
plt.show() 
```
